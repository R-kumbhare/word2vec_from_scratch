```python
import re
from nltk.corpus import  stopwords
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
import nltk
nltk.download('wordnet')
```

```python
corpus =[]
for i in range(0,len(df)):
    review = re.sub('[^a-zA-Z]'," ",df['messages'][i])
    review = review.lower()
    review = review.split()
    
    #review = [lemmatizer.lemmatize(word) for word in review if not word in stopwords]
    review = [lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')] 
    review = " ".join(review)
    corpus.append(review)
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
words1 =[]
words2 =[]
for sent in corpus:
    sent_token = sent_tokenize(sent)
    #print(sent_token)
    for sent in sent_token:
        words1.append(simple_preprocess(sent))
        

        
        
#print(words)
```

### Let's train word2vec from scratch

```python
import gensim
from gensim import models
from gensim.models import Word2Vec
model = gensim.models.Word2Vec(words1, window=5,min_count = 2)
```

### Average word2vec

```python
import numpy as np
```

```python
def avg_word2vec(doc):
    return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key],axis = 0)
```
